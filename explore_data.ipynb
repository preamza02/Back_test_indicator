```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import talib
```

```python
price_path = 'raw_data\coin_Bitcoin.csv'
full_data = pd.read_csv(price_path)
full_data['Date'] = pd.to_datetime(full_data['Date'])
full_data
```

```python
plt.figure(figsize = (20,10))
plt.plot(full_data['Date'],full_data['Close'])
plt.show()
```

```python
coin_name = 'BTC'
columns = 'Close'
start_date = '2020-01-01'
day_df = full_data['Date'][full_data['Date'] > start_date]
data = full_data[columns][full_data['Date'] > start_date]
EMA_low_date = 26
EMA_high_date = 12
EMA_low = talib.EMA(data,EMA_low_date)
EMA_high = talib.EMA(data,EMA_high_date)
plt.figure(figsize = (25,10))
plt.plot(day_df,data,color = 'blue',label = f'price')
plt.plot(day_df,EMA_low,color = 'green',label = f'EMA {EMA_low_date}')
plt.plot(day_df,EMA_high,color = 'red',label = f'EMA {EMA_high_date}')
plt.xlabel('Time Line')
plt.ylabel(f'{coin_name} Close Price USD ($)')
plt.legend()
plt.show()
```
